```python
import librosa
import io    
import matplotlib.pyplot as plt                  
import librosa.display
import numpy
# France, Italy, Spain, Switzerland, Germany , UK
```

```python
audio_data = 'Algeria.wav'                                                    
x,sr = librosa.load(audio_data) 
librosa.get_duration(x)
```

```python
audio_data2 = 'Kenya.wav'                                                    
x2,sr2 = librosa.load(audio_data2) 
librosa.get_duration(x2)
```

```python
audio_data3 = 'Tanzania.wav'                                                    
x3,sr3 = librosa.load(audio_data3) 
librosa.get_duration(x3)
```

```python
tempo, beat_times = librosa.beat.beat_track(x, sr=sr, start_bpm=60, units='time')
%matplotlib inline
# Plot the beat locations over the waveform:
plt.figure(figsize=(14, 5))
plt.title("Algeria")
librosa.display.waveplot(x, alpha=0.6)
plt.vlines(beat_times, -1, 1, color='r')
plt.ylim(-1, 1)

# Plot a histogram of the intervals between adjacent beats:
beat_times_diff = numpy.diff(beat_times)
plt.figure(figsize=(14, 5))
plt.hist(beat_times_diff, bins=50, range=(0,3))
plt.title("Algeria")
plt.xlabel('Beat Length (seconds)')
plt.ylabel('Count')
```
